```python
'''Adding a Series to a dataframe:'''
stadiums=Series(["Levi's Stadium","AT&T Stadium"],index=[8,0])
```

```python
stadiums
```

```python
'''Now, we call the dataFrame and its column to which we want to 
assign this Series'''
nfl_frame['Stadium']=stadiums
```

```python
'''
Python is very good at handling null values:
'''
nfl_frame
```

```python
'''we can also delete entire columns:'''
del nfl_frame['Stadium']
nfl_frame
```
